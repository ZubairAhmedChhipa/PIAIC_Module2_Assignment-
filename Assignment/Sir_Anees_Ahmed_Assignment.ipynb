```python
from PIL import Image
import os 
import glob
import shutil
import numpy as np
# for image reading
import matplotlib.pylab as plt
#import matplotlib.image



def resize_image(input_dir, infile, output_dir='resized', size=(200,200)):
    outfile = os.path.splitext(infile)[0] + '_resized'
    extension = os.path.splitext(infile)[1] 
    
    try:
        img = Image.open(input_dir + '/' + infile)
        img = img.resize((size[0], size[1]), Image.LANCZOS)
        
        new_file = output_dir + "/" + outfile + extension
        img.save(new_file)
        
    except IOError:
        print("unable to resize image {}".format(infile))

        

if __name__ == '__main__':
    output_dir = 'resized'
    dir = os.getcwd()
    input_dir = 'images'
    full_input_dir = dir + '/' + input_dir
    
    
    if not os.path.exists(os.path.join(dir, output_dir)):
        os.mkdir(output_dir)
    
    try:
        for file in os.listdir(full_input_dir):
            resize_image(input_dir, file, output_dir)
    except OSError:
        print('file not found')
        
 




src_dir = r"C:\Users\Chhipa\AppData\Local\Programs\Python\Python37-32\Scripts\Assignment\resized"
dst_dir = r"C:\Users\Chhipa\AppData\Local\Programs\Python\Python37-32\Scripts\Assignment"
for jpgfile in glob.iglob(os.path.join(src_dir, "*.jpg")):
    shutil.copy(jpgfile, dst_dir)

    
def plti(im, h=8, **kwargs):
    """
    Helper function to plot an image.
    """
   
    y = im.shape[0]
    x = im.shape[1]
    w = (y/x) * h
    plt.figure(figsize=(w,h))
    plt.imshow(im, interpolation="none", **kwargs)
    plt.axis('off')
    #input('press <ENTER> to continue')
    
    
    
    
#%matplotlib pltinline



a = np.array([([plt.imread("1_resized.jpg")], [plt.imread("2_resized.jpg")], [plt.imread("3_resized.jpg")],
          [plt.imread("4_resized.jpg")],[plt.imread("5_resized.jpg")],[plt.imread("6_resized.jpg")],
          [plt.imread("7_resized.jpg")],[plt.imread("8_resized.jpg")],[plt.imread("9_resized.jpg")],
          [plt.imread("10_resized.jpg")],[plt.imread("11_resized.jpg")],[plt.imread("12_resized.jpg")],
          [plt.imread("13_resized.jpg")], [plt.imread("14_resized.jpg")],[plt.imread("15_resized.jpg")],
          [plt.imread("16_resized.jpg")],[plt.imread("17_resized.jpg")],[plt.imread("18_resized.jpg")],
          [plt.imread("19_resized.jpg")],[plt.imread("20_resized.jpg")])])


print(np.shape(a))
print(type(np.array(a)))
np.array(a)
#plt.imshow(np.array(plti(a)))

#print(plti(a))
#np.array([[plti(im1)], [plti(im2)], [plti(im3)], [plti(im4)], [plti(im5)], [plti(im6)], [plti(im7)], [plti(im8)], [plti(im9)], [plti(im10)], [plti(im11)], [plti(im12)], [plti(im13)], [plti(im14)], [plti(im15)], [plti(im16)], [plti(im17)], [plti(im18)], [plti(im19)], [plti(im20)]])

#print(np.shape([[im1], [im2], [im3], [im4], [im5], [im6], [im7], [im8], [im9], [im10], [im11], [im12], [im13], [im14], [im15], [im16], [im17], [im18], [im19], [im20]]))

#print(type(np.array([[im1], [im2], [im3], [im4], [im5], [im6], [im7], [im8], [im9], [im10], [im11], [im12], [im13], [im14], [im15], [im16], [im17], [im18], [im19], [im20]])))

#print(np.array([[im1], [im2], [im3], [im4], [im5], [im6], [im7], [im8], [im9], [im10], [im11], [im12], [im13], [im14], [im15], [im16], [im17], [im18], [im19], [im20]]))

    
```
